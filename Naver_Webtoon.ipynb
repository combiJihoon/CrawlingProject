```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://comic.naver.com/webtoon/weekday.nhn")
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
webtoon_by_day = soup.select('.col_inner .title')
```

```python
webtoon_by_day
```

```python
webtoon_list = []
for webtoon in webtoon_by_day :
    webtoon_list.append(webtoon.text.strip())
```

```python
webtoon_list
```

```python
days_tag = soup.select('.thumb + .title')
```

```python
days_tag
```

```python
webtoon_title_list = []
for webtoon in days_tag :
    webtoon_title_list.append(webtoon.text)
```

```python
webtoon_title_list
```

월요일일 경우 : col_inner 안에 ".mon" 이 있고 그 아래로는 li들이 있는데 여기에 웹툰 제목이 들어 있다. 
즉, col_inner가 자식 선택자로 ".mon"를 가질 경우, col_inner의 또다른 자식 선택자인 ".title" 선택자를 선택해라
